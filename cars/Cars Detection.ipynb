```python
# Importando a biblioteca do OpenCV
import cv2

# Importando biblioteca para exibir as imagens
from matplotlib import pyplot as plt

# Importando biblioteca para manipular imagens
import numpy as np

from math import sqrt

# Carregando a imagem
cars = cv2.imread('images/cars.jpg') # Lendo a imagem
cars = cv2.cvtColor(cars, cv2.COLOR_BGR2RGB) # Convertendo a imagem de BGR para RGB

cars2 = cv2.imread('images/cars2.jpg') # Lendo a imagem
cars2 = cv2.cvtColor(cars2, cv2.COLOR_BGR2RGB) # Convertendo a imagem de BGR para RGB

car_gray = cv2.cvtColor(cars, cv2.COLOR_BGR2GRAY)
car_gray2 = cv2.cvtColor(cars2, cv2.COLOR_BGR2GRAY)


# Exibindo a imagem
plt.imshow(cars)
plt.show()
plt.imshow(cars2)
plt.show()
```

```python
# Carregando o classificador

classifier_cars = cv2.CascadeClassifier('classifiers/cars.xml')
classifier_cars2 = cv2.CascadeClassifier('classifiers/cars.xml')
```

```python
detect_cars = classifier_cars.detectMultiScale(car_gray,1.1,1)
detect_cars2 = classifier_cars.detectMultiScale(car_gray2,1.1,1)

car_cp = cars.copy()
car_cp2 = cars2.copy()

# Desenhando os retângulos nos carros detectados pelo classificador

for (x,y,w,h) in detect_cars:
    cv2.rectangle(car_cp,(x,y),(x+w,y+h),(255,0,0),2)

    
for (x,y,w,h) in detect_cars2:
    cv2.rectangle(car_cp2,(x,y),(x+w,y+h),(255,255,255),2)
    
    
    
dpi = 100
height, width = cars.shape[:2]

figsize = width / dpi, height / dpi

# Criando a figura com as características acima
fig = plt.figure(figsize=figsize)    

# Exibindo o resultado
plt.imshow(car_cp)
plt.axis("off")
plt.show()

plt.imshow(car_cp2)
plt.axis("off")
plt.show()
```
