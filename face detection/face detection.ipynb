```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from math import sqrt
```

```python
# Carregando o arquivo
faces = cv2.imread('images/face.jpg')
faces = cv2.cvtColor(faces, cv2.COLOR_BGR2RGB)

faces2 = cv2.imread('images/face2.jpg')
faces2 = cv2.cvtColor(faces2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(faces)
plt.show()
```

```python
plt.imshow(faces2)
plt.show()
```

```python
classifier_face = cv2.CascadeClassifier('classifiers/haarcascade_frontalface_alt.xml')
detect_face = classifier_face.detectMultiScale(faces,1.1,1)
```

```python
face_copy = faces.copy()

for (x,y,w,h) in detect_face:
    cv2.rectangle(face_copy,(x,y),(x+w,y+h),(255,0,0),8)
```

```python
plt.imshow(face_copy)
plt.show()
```

```python
classifier_face2 = cv2.CascadeClassifier('classifiers/haarcascade_frontalface_alt.xml')
detect_face2 = classifier_face2.detectMultiScale(faces2,1.1,1)
```

```python
face_copy2 = faces2.copy()

for (x,y,w,h) in detect_face2:
    cv2.rectangle(face_copy2,(x,y),(x+w,y+h),(255,255,0),2)
```

```python
plt.imshow(face_copy2)
plt.show()
```

Cortando as faces da imagem original depois de descobrir os pontos na imagem

```python
crop_face = []

for (x,y,w,h) in detect_face:
    crop_face.append(faces[y:y+h, x:x+w])

for f in crop_face:
    plt.figure()
    plt.imshow(f)
    plt.show()
    
```
